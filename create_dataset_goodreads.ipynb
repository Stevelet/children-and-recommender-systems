**Specify your directory here:**

```python
DIR = './'
file_ids = pd.read_csv(os.path.join(DIR, 'gdrive_id.csv'))
display(file_ids)
file_id_map = dict(zip(file_ids['name'].values, file_ids['id'].values))

def download_by_name(fname, output=None, quiet=False):
    if fname in file_id_map:
        url = 'https://drive.google.com/uc?id='+file_id_map[fname]
        gdown.download(url, output=output, quiet=quiet)
    else:
        print('The file', fname, 'can not be found!')

download_by_name('goodreads_interactions_children.json.gz')
download_by_name('goodreads_books_children.json.gz')
download_by_name('goodreads_reviews_children.json.gz')
```

**This function shows how to load datasets**

```python
def load_data(file_name, head = None):
    count = 0
    data = []
    with gzip.open(file_name) as fin:
        for l in fin:
            d = json.loads(l)
            count += 1
            data.append(d)
            
            # break if reaches the 100th line
            if (head is not None) and (count > head):
                break
    return data
```

**Load and display sample records of books/authors/works/series**

```python
books = load_data(os.path.join(DIR, 'goodreads_books_children.json.gz'))
df = pd.DataFrame(books)
df.to_csv("result.csv")
df
```
